```python

# 1.
def print_first_10():
    for i in range(1, 11):
        print(i, end=" ")
    print()
print_first_10()
```

```python


# 2. 
def sum_natural(n):
    return n * (n + 1) // 2
```

```python
sum_natural(10)
```

```python


# 3.
def reverse_number(num):
    rev = 0
    while num > 0:
        rev = rev * 10 + num % 10
        num //= 10
    return rev


```

```python
reverse_number(12345)
```

```python
# 4.
def count_digits(num):
    if num == 0:
        return 1
    count = 0
    num = abs(num)
    while num > 0:
        count += 1
        num //= 10
    return count
print (count_digits(6666675))
```

```python
# 5.
def is_palindrome(num):
    return num == reverse_number(num)
print(is_palindrome(43534))
```

```python
is_palindrome(46734)
```

```python
# 6. 
def fibonacci(n):
    series = []
    a, b = 0, 19
    for _ in range(n):
        series.append(a)
        a, b = b, a + b
    return series
print(fibonacci(6))
```

```python
# 7. 
def add(a, b): return a + b
def sub(a, b): return a - b
def mul(a, b): return a * b
def div(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Division by zero!"

def calculator():
    print("\n--- Calculator ---")
    print("1. Add 2. Subtract 3. Multiply 4. Divide")
    choice = int(input("Select operation: "))
    x = float(input("Enter first number: "))
    y = float(input("Enter second number: "))
    
    if choice == 1: print("Result:", add(x, y))
    elif choice == 2: print("Result:", sub(x, y))
    elif choice == 3: print("Result:", mul(x, y))
    elif choice == 4: print("Result:", div(x, y))
    else: print("Invalid choice")

```

```python
calculator()
```

```python
# 8.
def write_student_file(name, marks):
    with open("student.txt", "w") as f:
        f.write(f"Name: {name}\nMarks: {marks}")
    print("Saved to student.txt")
```

```python
write_student_file("avni", 99)
```

```python
# 9.
def read_student_file():
    try:
        with open("student.txt", "r") as f:
            print("\nFile content:")
            print(f.read())
    except FileNotFoundError:
        print("File not found!")
        
```

```python
read_student_file()
```

```python
# 10.
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    def display(self):
        print(f"Name: {self.name}, Marks: {self.marks}")

print("1. First 10 natural numbers:")
print_first_10()

print("\n2. Sum of first 10 natural numbers:", sum_natural(10))

print("\n3. Reverse of 12345:", reverse_number(12345))

print("\n4. Digits in 987654:", count_digits(987654))

print("\n5. Is 121 palindrome?", is_palindrome(121))
print("   Is 123 palindrome?", is_palindrome(123))

print("\n6. First 10 Fibonacci numbers:", fibonacci(10))

print("\n7. Calculator test:")
print("5 + 3 =", add(5, 3))
print("10 / 2 =", div(10, 2))
print("10 / 0 =", div(10, 0))

print("\n8 & 9. File handling test:")
write_student_file("Aman", 88)
read_student_file()

print("\n10. Student Object:")
s = Student("Rahul", 95)
s.display()
```
